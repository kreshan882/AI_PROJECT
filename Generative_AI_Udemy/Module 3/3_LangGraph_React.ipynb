```python
from langgraph.prebuilt import create_react_agent


@tool
def get_weather(city: str) -> str:
    """Get weather for a given city."""
    res_from3rdParty="today (9.00am- 11.00am) rain in", {city}
    return res_from3rdParty

# model = ChatOpenAI(model="gpt-4o", temperature=0.6)

agent = create_react_agent(
    model="openai:gpt-4.1",
    tools=[get_weather],
    prompt="Give me the simple anser, in Tamil"
)

# Run the agent
response=agent.invoke(
    {"messages": [{"role": "user", "content": "what is the weather in Colombo"}]}
)

print(response["messages"][-1].content)
```
