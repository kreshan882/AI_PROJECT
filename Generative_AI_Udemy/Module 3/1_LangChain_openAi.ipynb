# Chat History

```python
chat_history = [
    SystemMessage(content="give me the simple anser, u are the tamil assistent")
]

first_question = HumanMessage(content="what is the Capital in Srilanka?")

chat_history.append(first_question)

first_response = llm.invoke(chat_history)
chat_history.append(first_response)


second_question = HumanMessage(content="what was my prev question?")
chat_history.append(second_question)

second_response = llm.invoke(chat_history)
chat_history.append(second_response)
```

# Chain

```python
from langchain.prompts import ChatPromptTemplate

template = ChatPromptTemplate.from_template("tel me the people count in district: {topic},"
" in srilanka, simple anser only")

chain = template | llm

response = chain.invoke({"topic": "colombo"})

print(response.content)
```
