```python
import os
from langchain_openai import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage, AIMessage


os.environ["OPENAI_API_KEY"] = "your-api-key"

llm = ChatOpenAI(model="gpt-4o")

print("LLM Initialized:", llm.model_name)
```

```python
system_instruction = SystemMessage(content="You are a helpful assistant that translates English to tamil.")

user_question = HumanMessage(content="I love programming.")
```

```python

response = llm.invoke([system_instruction, user_question])

print("AI Response:")
print(response.content)
```

# Chat History

```python
chat_history = [
    SystemMessage(content="You are a witty chatbot that knows a lot about space.")


]

first_question = HumanMessage(content="What is the largest planet in our solar system?")

chat_history.append(first_question)

first_response = llm.invoke(chat_history)
```

```python
first_response
```

```python
chat_history.append(first_response)

print(f"Human: {first_question.content}")
print(f"AI: {first_response.content}")
```

```python
second_question = HumanMessage(content="what was my prev question?")
chat_history.append(second_question)

second_response = llm.invoke(chat_history)
chat_history.append(second_response)

print(f"\nHuman: {second_question.content}")
print(f"AI: {second_response.content}")
```

# Chain

```python
from langchain.prompts import ChatPromptTemplate

template = ChatPromptTemplate.from_template("Tell me a joke about {topic}.")

chain = template | llm

response = chain.invoke({"topic": "computers"})

print(response.content)
```
