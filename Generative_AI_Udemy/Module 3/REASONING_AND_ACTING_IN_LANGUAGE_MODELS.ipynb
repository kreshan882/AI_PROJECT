```python
from openai import OpenAI

api_key="your-api-key"

client = OpenAI(api_key=api_key)

def get_response(system_msg, user_msg):
    msgs=[
        {"role": "system", "content": system_msg},
        {"role": "user", "content": user_msg},

    ]

    response = client.chat.completions.create(
    model="gpt-4o",
    messages=msgs,
    stop=["Observation:"],
    temperature=0.0,
    )

    return response.choices[0].message.content
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()

def search_duck(query):
  result = search_tool.invoke(query)
  return result
```

```python

system_msg = """You are a helpful assistant.
Available tools:
SearchDocs(query): search documentation and examples

* Always reason step by step using Thought → Action → Observation → Finish before giving the final answer.

If you do not know the answer:

Thought: explain what you are thinking and what information you need to find.

Action: write the exact search query or tool action you would perform to find the answer.

Observation: describe what you learn from the search or tool.

Finish: summarize the answer clearly, even if it is based on what you found.

Do not skip any of these steps, even if you are unsure of the answer. Always show your reasoning process.
"""
user_msg="""
Example 1-shot:

User: What is the new iPhone?

Assistant:

Thought: I need to find out the latest iPhone model and its key features.

Action: SearchDocs('latest iPhone 2025')

Observation: The latest iPhone is the iPhone 16 Pro, released in September 2025, featuring an improved camera system, A19 chip, and longer battery life.

Finish: The new iPhone is the iPhone 16 Pro (2025 release), which includes an upgraded camera, the A19 chip, and extended battery life.

Example 2-shot:

User: What is 456 times 78?

Assistant:

Thought: The user is asking for a multiplication calculation. I should use the `Calculator` tool.

Action: Calculator(456, 78)

Observation: 35568

Finish: The result of 456 times 78 is 35,568.

Example 3-shot:

User: How do I use the useState hook in React?

Assistant:

Thought: The user is asking for information about a specific concept in React. I should use the `SearchDocs` tool to find documentation and examples.

Action: SearchDocs('how to use useState hook React example')

Observation: The `useState` hook allows you to add state to function components. It is imported from React and called with an initial state value, returning an array with the current state value and a function to update it. Example: `const [count, setCount] = useState(0);`

Finish: You use the `useState` hook in React to add state to a functional component. You import it from React, and call it with an initial value. It returns a pair of values: the current state and a function that updates it (e.g., `const [value, setValue] = useState(initialValue);`).

User:What happened recetly in vijay rally at karur ?
Assistant:"""

response =get_response(system_msg, user_msg)
print(response)
```

```python
import re
def extract_and_execute_action(response_text):
    """
    Extracts the tool action and query from the AI's response text and executes it.
    """
    action_line_match = re.search(r"Action: (.+)", response_text)

    if not action_line_match:
        print("Error: 'Action:' line not found in the response.")
        return None

    action_text = action_line_match.group(1).strip()

    tool_call_match = re.search(r"(\w+)\(['\"](.+?)['\"]\)", action_text)

    if not tool_call_match:
        print(f"Error: Could not parse tool call from: {action_text}")
        return None

    tool_name = tool_call_match.group(1)
    query = tool_call_match.group(2)

    print(f"Extracted Tool: {tool_name}")
    print(f"Extracted Query: {query}")

    if tool_name == "SearchDocs":
        result = search_duck(query)
        return result
    else:
        print(f"Unknown tool: {tool_name}")
        return None
```

```python
search_data=extract_and_execute_action(response)
print(search_data)
```

```python
new_user_msg=user_msg + "\n" + response + "Observation:\n"+search_data + "\n\n" + "Finish:"
print(new_user_msg)
```

```python
response =get_response(system_msg, new_user_msg)
print(response)
```
