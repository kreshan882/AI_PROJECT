# Stop parameter

````python
user_msg = """
Tell me a very short bedtime story.
"""

msgs = [
    {"role": "user", "content": user_msg},
]

response = client.chat.completions.create(
    model="gpt-4o",
    messages=msgs,
    stop=["the"],
    temperature=0.0
)

print(response.choices[0].message.content)
````

# Streaming the response

````python
from openai import OpenAI

api_key=""
client = OpenAI(api_key=api_key)

user_msg = """
Tell me a very short bedtime story.
"""
msgs = [
    {"role": "user", "content": user_msg},
]

response = client.chat.completions.create(
    model="gpt-4o",
    messages=msgs,
    temperature=0.0,
    stream=True
)

# Iterate over the stream
for chunk in response:
    if chunk.choices[0].delta.content is not None:
      data=chunk.choices[0].delta.content
      data=data.strip()
      if data.lower()=="in":
        break
      print(chunk.choices[0].delta.content, end="", flush=True)
````

# **Prompt Engineering**

* Instruction - a specific task or instruction you want the model to perform

* Context - external information or additional context that can steer the model to better responses

* Input Data - the input or question that we are interested to find a response for

- https://www.promptingguide.ai/introduction/elements



```
Instruction:
Answer the question based on the context provided. Give a concise and accurate response.

Context:
"Alice is a 28-year-old software engineer who lives in San Francisco. She enjoys hiking on weekends, cooking Italian cuisine, and reading mystery novels. She has a younger brother named Tom who works as a graphic designer. Alice recently adopted a dog named Max and volunteers at a local animal shelter."

Input Data (Question):
What are Alice’s hobbies?
```



````python
def get_response(user_msg):
    msgs=[
        {"role": "user", "content": user_msg},

    ]

    response = client.chat.completions.create(
    model="gpt-5-nano",
    messages=msgs
    )

    return response.choices[0].message.content
````

# Zero-shot (likely to fail the format)

````python
user_msg = """
What is the capital of Japan? Give the answer in JSON format.
"""

print(get_response(user_msg))
````

# Few-shot (shows pattern → model learns it)

````python
user_msg = """
You are a helpful assistant. Always answer in JSON format.

Examples:

Q: What is the capital of France?
A: {
  "answer": "Paris"
}

Q: Who discovered gravity?
A: {
  "answer": "Isaac Newton"
}

Now, Q: What is the capital of Japan?
"""

print(get_response(user_msg))
````

**For simple task zero shot or one shot is okay , but when problem is very complex also LLM model is not responding as we expect we give me them more shots(examples)**
