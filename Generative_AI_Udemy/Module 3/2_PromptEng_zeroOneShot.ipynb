# Streaming the response

```python


user_msg = "Tell me a very short bedtime story." 

msgs = [
    {"role": "user", "content": user_msg},
]

response = client.chat.completions.create(
    model="gpt-4o",
    messages=msgs,
    temperature=0.6,
    stream=True
)

# Iterate over the stream
for chunk in response:
    if chunk.choices[0].delta.content is not None:
      data=chunk.choices[0].delta.content
      if data.strip().lower()=="the":
        print()
      print(data, end="", flush=True)

```

# **Prompt Engineering**

* Instruction - a specific task or instruction you want the model to perform

* Context - external information or additional context that can steer the model to better responses

* Input Data - the input or question that we are interested to find a response for





```python
def get_response(user_msg):
    msgs=[
        {"role": "user", "content": user_msg},

    ]

    response = client.chat.completions.create(
    model="gpt-5-nano",
    messages=msgs
    )

    return response.choices[0].message.content
```

# Zero-shot (likely to fail the format)

```python
user_msg = """
What is the capital of Japan? Give the answer in JSON format.
"""

print(get_response(user_msg))
```

# Few-shot (shows pattern → model learns it)

```python
user_msg = """
You are a helpful assistant. Always answer in JSON format.

Examples:

Q: What is the capital of France?
A: {
  "answer": "Paris"
}

Q: Who discovered gravity?
A: {
  "answer": "Isaac Newton"
}

Now, Q: What is the capital of Japan?
"""

print(get_response(user_msg))
```
