```python
import numpy as np

def is_normalized(vector, tol=1e-6):
    norm = np.linalg.norm(vector)
    return abs(norm - 1.0) < tol
```

```python
print(is_normalized(response.data[0].embedding))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def cosine_sim_sklearn(vec1, vec2):
    # The library expects 2D arrays, so we reshape the vectors
    vector1_reshaped = [vec1]
    vector2_reshaped = [vec2]

    # The result is a 2D array (matrix), so we access the value at [0, 0]
    return cosine_similarity(vector1_reshaped, vector2_reshaped)[0][0]
```

```python
def openai_embed(query):
  response = client.embeddings.create(
    input=query,
    model="text-embedding-3-small"
  )

  return response.data[0].embedding
```

```python
cosine_sim_sklearn(openai_embed("what is my hr mail Id?"),openai_embed("SAP IS USED FOR SOME AUTOMATION in finances"))
```

```python
cosine_sim_sklearn(openai_embed("what is my hr mail Id?"),openai_embed("HR mail I d is hr@hr.com"))
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

def sentence_transformer_embed(text):

    embedding = model.encode(text)
    return embedding
```

```python
cosine_sim_sklearn(sentence_transformer_embed("python programming"),sentence_transformer_embed("python snake"))
```
