```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load iris dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

df.head()
```

```python
# Plot pairwise scatter plots (4 vs 4)
sns.pairplot(df, hue='species', diag_kind='kde', palette='husl')
plt.suptitle("Iris Dataset: 4 Features vs 4 Features", y=1.02)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create simple example data
data = {
    "Customer": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Clothes_Spend": [200, 220, 30, 40, 210, 35, 45, 230],
    "Food_Spend": [50, 60, 180, 170, 55, 160, 175, 40]
}

df = pd.DataFrame(data)


# Plot
plt.figure(figsize=(6, 4))
plt.scatter(df["Clothes_Spend"], df["Food_Spend"], color="skyblue", s=100, edgecolor='black')

# Labels for customers
for i, row in df.iterrows():
    plt.text(row["Clothes_Spend"] + 2, row["Food_Spend"] + 2, row["Customer"])

plt.xlabel("Spending on Clothes ($)")
plt.ylabel("Spending on Food ($)")
plt.title("Unsupervised Learning: Seeing Natural Groups (No Algorithm)")
plt.grid(True)
plt.show()
```

```python
# Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import pandas as pd
import joblib  # For saving and loading models

# Load dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Split features and target
X = df[iris.feature_names]
y = df['species']

# Train the Decision Tree
model = DecisionTreeClassifier(max_depth=5, random_state=42)
model.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(16, 10))
plot_tree(
    model,
    feature_names=iris.feature_names,
    class_names=iris.target_names,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree on Iris Dataset")
plt.show()

```

```python
# Save the model
joblib.dump(model, "iris_decision_tree.pkl")
print("Model saved as iris_decision_tree.pkl")
```

```python
import warnings
# -------------------------------
# Example: Sending input values for prediction
# -------------------------------

new_sample = [[2.1, 3.6, 2.4, 0.3]]

warnings.filterwarnings(action='ignore', category=UserWarning, module='sklearn') 

predicted_class_index = model.predict(new_sample)[0]
predicted_class_name = iris.target_names[predicted_class_index]

print(f"Predicted class index: {predicted_class_index}")
print(f"Predicted class name: {predicted_class_name}")
```

```python

# -------------------------------
# Optional: Load the model later and predict
# -------------------------------
loaded_model = joblib.load("iris_decision_tree.pkl")
print("Model loaded from disk.")
prediction = loaded_model.predict(new_sample)
print("Prediction from loaded model:", iris.target_names[prediction[0]])
```
