```python
from nltk.translate.bleu_score import sentence_bleu

candidate = "the cat is on the".split()
reference = ["the cat is on the floor".split()]

# BLEU
bleu = sentence_bleu(reference, candidate)
print("BLEU:", bleu)
```

```python
from rouge_score import rouge_scorer


# ROUGE
scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
scores = scorer.score("the cat is on the mat", "the cat is on the floor")
print("ROUGE:", scores)
```

```python

candidate = "the cat is on the mat".split()
reference = ["A feline is resting on the floor covering.".split()]

# BLEU
bleu = sentence_bleu(reference, candidate)
print("BLEU:", bleu)
# ROUGE
scorer = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
scores = scorer.score("the cat is on the mat", "A feline is resting on the floor covering.")
print("ROUGE:", scores)
```

```python
from bert_score import score

reference = ["the cat is on the mat"]
candidate = ["the cat is on the floor"]

P, R, F1 = score(candidate, reference, lang="en", model_type="roberta-large",verbose=True)

print(f"Precision: {P.mean().item():.4f}")
print(f"Recall:    {R.mean().item():.4f}")
print(f"F1:        {F1.mean().item():.4f}")
```
