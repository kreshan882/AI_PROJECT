```python
import os
from openai import OpenAI
import json

os.environ["OPENAI_API_KEY"] = "your-api-key"
client = OpenAI()
```

```python

question = "What is the capital of France and what is it famous for?"

ground_truth = "The capital of France is Paris. It is famous for the Eiffel Tower, Louvre Museum, art, culture, cuisine, and fashion."

correct_answer = """
Paris is the capital of France. The city is famous for its iconic landmarks
including the Eiffel Tower and the Louvre Museum. Paris is also renowned for
its rich cultural heritage, world-class cuisine, and influence on fashion and art.
"""

irrelevant_answer = """
France is a country in Europe. Europe has many countries with different languages.
The weather in Europe varies by region. Many tourists visit Europe every year
for various reasons including sightseeing and cultural experiences.
"""
```

# Answer Relevancy

```python


def evaluate_answer_relevancy(question, answer):
    """
    Evaluates whether the answer addresses the question in an informative
    and concise manner.
    """
    prompt = f"""
You are an expert evaluator. Your task is to evaluate the relevancy of an answer to a given question.

Evaluation Criteria:
- Does the answer directly address the question?
- Is the answer informative and complete?
- Is the answer concise without unnecessary information?

Question: {question}

Answer: {answer}

Please evaluate the answer on a scale of 1-5 where:
1 = Completely irrelevant, doesn't address the question
2 = Minimally relevant, barely addresses the question
3 = Somewhat relevant, partially addresses the question
4 = Mostly relevant, addresses most aspects of the question
5 = Highly relevant, fully addresses the question in an informative and concise manner

Provide your evaluation in the following JSON format:
{{
    "score": <1-5>,
    "reasoning": "<explanation for your score>"
}}
"""

    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    result = json.loads(response.choices[0].message.content)
    return result

print("Answer Relevancy evaluator defined!")

print("=" * 80)
print("EVALUATION 1: ANSWER RELEVANCY")
print("=" * 80)

# Evaluate correct answer
print("\n--- Evaluating CORRECT Answer ---")
relevancy_correct = evaluate_answer_relevancy(question, correct_answer)
print(f"Score: {relevancy_correct['score']}/5")
print(f"Reasoning: {relevancy_correct['reasoning']}")

# Evaluate irrelevant answer
print("\n--- Evaluating IRRELEVANT Answer ---")
relevancy_irrelevant = evaluate_answer_relevancy(question, irrelevant_answer)
print(f"Score: {relevancy_irrelevant['score']}/5")
print(f"Reasoning: {relevancy_irrelevant['reasoning']}")
```

# Coherence (the quality of being logical and consistent.)

```python
def evaluate_coherence(answer):
    """
    Evaluates the coherence and logical flow of the answer.
    """
    prompt = f"""
You are an expert evaluator. Your task is to evaluate the coherence of an answer.

Evaluation Criteria:
- Is the answer well-structured and organized?
- Do the sentences flow logically from one to another?
- Is the answer easy to understand?
- Are ideas connected clearly?

Answer: {answer}

Please evaluate the coherence on a scale of 1-5 where:
1 = Completely incoherent, no logical structure
2 = Minimally coherent, poor structure and flow
3 = Somewhat coherent, basic structure with some flow issues
4 = Mostly coherent, good structure with minor flow issues
5 = Highly coherent, excellent structure and logical flow

Provide your evaluation in the following JSON format:
{{
    "score": <1-5>,
    "reasoning": "<explanation for your score>"
}}
"""

    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    result = json.loads(response.choices[0].message.content)
    return result

print("Coherence evaluator defined!")

print("\n" + "=" * 80)
print("EVALUATION 2: COHERENCE")
print("=" * 80)

# Evaluate correct answer
print("\n--- Evaluating CORRECT Answer ---")
coherence_correct = evaluate_coherence(correct_answer)
print(f"Score: {coherence_correct['score']}/5")
print(f"Reasoning: {coherence_correct['reasoning']}")

# Evaluate irrelevant answer
print("\n--- Evaluating IRRELEVANT Answer ---")
coherence_irrelevant = evaluate_coherence(irrelevant_answer)
print(f"Score: {coherence_irrelevant['score']}/5")
print(f"Reasoning: {coherence_irrelevant['reasoning']}")
```

# Correctness (Factual Accuracy)

```python

def evaluate_correctness(question, answer, ground_truth):
    """
    Evaluates whether the answer is factually correct based on ground truth.
    """
    prompt = f"""
You are an expert evaluator. Your task is to evaluate the factual correctness of an answer.

Evaluation Criteria:
- Are the facts in the answer accurate according to the ground truth?
- Does the answer contain any false or misleading information?
- Does the answer align with the ground truth?

Question: {question}

Ground Truth: {ground_truth}

Answer to Evaluate: {answer}

Please evaluate the correctness on a scale of 1-5 where:
1 = Completely incorrect, contains false information
2 = Mostly incorrect, many factual errors
3 = Partially correct, some accurate facts mixed with errors
4 = Mostly correct, minor inaccuracies
5 = Completely correct, all facts are accurate

Provide your evaluation in the following JSON format:
{{
    "score": <1-5>,
    "reasoning": "<explanation for your score>"
}}
"""

    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    result = json.loads(response.choices[0].message.content)
    return result

print("Correctness evaluator defined!")

print("\n" + "=" * 80)
print("EVALUATION 3: CORRECTNESS (Factual Accuracy)")
print("=" * 80)

# Evaluate correct answer
print("\n--- Evaluating CORRECT Answer ---")
correctness_correct = evaluate_correctness(question, correct_answer, ground_truth)
print(f"Score: {correctness_correct['score']}/5")
print(f"Reasoning: {correctness_correct['reasoning']}")

# Evaluate irrelevant answer
print("\n--- Evaluating IRRELEVANT Answer ---")
correctness_irrelevant = evaluate_correctness(question, irrelevant_answer, ground_truth)
print(f"Score: {correctness_irrelevant['score']}/5")
print(f"Reasoning: {correctness_irrelevant['reasoning']}")
```

# Context Relevancy Score


```python

query = "What is the capital of France?"

# Scenario 1: ALL RELEVANT CONTEXT (Best case)
all_relevant_context = [
    "Paris is the capital and largest city of France.",
    "Paris is located in northern France on both banks of the Seine River.",
    "The city of Paris is the capital of France and has been for centuries.",
    "Paris serves as the political and administrative center of France."
]

# Scenario 2: MIX OF RELEVANT AND IRRELEVANT (Medium case)
mixed_context = [
    "Paris is the capital and largest city of France.",  # Relevant
    "The Eiffel Tower is made of iron and was built in 1889.",  # Somewhat relevant
    "Spain is a country in Southern Europe with Madrid as its capital.",  # Irrelevant
    "France is known for its wine and cheese production."  # Somewhat relevant
]

# Scenario 3: ALL IRRELEVANT CONTEXT (Worst case)
all_irrelevant_context = [
    "The weather today is sunny and warm.",
    "Python is a popular programming language.",
    "Soccer is the most popular sport in the world.",
    "Coffee is grown in tropical regions."
]

print("Query:", query)
print("\nContexts defined for 3 scenarios:")
print("1. All Relevant")
print("2. Mixed (Relevant + Irrelevant)")
print("3. All Irrelevant")


def evaluate_rag_context(query, retrieved_contexts):
    """
    Evaluates if the retrieved contexts are relevant to answer the query.
    This simulates evaluating a RAG retriever's performance.
    """

    # Join contexts into a single string
    context_text = "\n".join([f"{i+1}. {ctx}" for i, ctx in enumerate(retrieved_contexts)])

    prompt = f"""
You are an expert evaluator for RAG (Retrieval-Augmented Generation) systems.
Your task is to evaluate if the retrieved contexts contain relevant information to answer the query.

Query: {query}

Retrieved Contexts:
{context_text}

Evaluation Criteria:
- Do the contexts contain information needed to answer the query?
- How much of the retrieved information is relevant vs irrelevant?
- Can the query be answered using these contexts?

Please evaluate on a scale of 1-5 where:
1 = No relevant information, completely useless contexts
2 = Minimal relevant information, mostly irrelevant
3 = Some relevant information mixed with irrelevant content
4 = Mostly relevant information, minor irrelevant content
5 = Highly relevant information, all contexts are useful

Also provide:
- Number of relevant contexts out of total
- Relevancy percentage

Provide your evaluation in JSON format:
{{
    "score": <1-5>,
    "relevant_contexts": <number>,
    "total_contexts": <number>,
    "relevancy_percentage": <0-100>,
    "reasoning": "<explanation>"
}}
"""

    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        temperature=0
    )

    result = json.loads(response.choices[0].message.content)
    return result

print("RAG Context Evaluator defined!")


print("\n" + "=" * 80)
print("SCENARIO 1: ALL RELEVANT CONTEXT")
print("=" * 80)
print(f"\nQuery: {query}")
print("\nRetrieved Contexts:")
for i, ctx in enumerate(all_relevant_context, 1):
    print(f"  {i}. {ctx}")

print("\n--- Evaluation ---")
eval_1 = evaluate_rag_context(query, all_relevant_context)
print(f"Score: {eval_1['score']}/5")
print(f"Relevant Contexts: {eval_1['relevant_contexts']}/{eval_1['total_contexts']}")
print(f"Relevancy: {eval_1['relevancy_percentage']}%")
print(f"Reasoning: {eval_1['reasoning']}")


print("\n" + "=" * 80)
print("SCENARIO 2: MIXED CONTEXT (Relevant + Irrelevant)")
print("=" * 80)
print(f"\nQuery: {query}")
print("\nRetrieved Contexts:")
for i, ctx in enumerate(mixed_context, 1):
    print(f"  {i}. {ctx}")

print("\n--- Evaluation ---")
eval_2 = evaluate_rag_context(query, mixed_context)
print(f"Score: {eval_2['score']}/5")
print(f"Relevant Contexts: {eval_2['relevant_contexts']}/{eval_2['total_contexts']}")
print(f"Relevancy: {eval_2['relevancy_percentage']}%")
print(f"Reasoning: {eval_2['reasoning']}")


print("\n" + "=" * 80)
print("SCENARIO 3: ALL IRRELEVANT CONTEXT")
print("=" * 80)
print(f"\nQuery: {query}")
print("\nRetrieved Contexts:")
for i, ctx in enumerate(all_irrelevant_context, 1):
    print(f"  {i}. {ctx}")

print("\n--- Evaluation ---")
eval_3 = evaluate_rag_context(query, all_irrelevant_context)
print(f"Score: {eval_3['score']}/5")
print(f"Relevant Contexts: {eval_3['relevant_contexts']}/{eval_3['total_contexts']}")
print(f"Relevancy: {eval_3['relevancy_percentage']}%")
print(f"Reasoning: {eval_3['reasoning']}")


print("\n" + "=" * 80)
print("COMPARISON SUMMARY")
print("=" * 80)

comparison_data = [
    ("All Relevant", eval_1),
    ("Mixed", eval_2),
    ("All Irrelevant", eval_3)
]

print(f"\n{'Scenario':<20} {'Score':<10} {'Relevant':<15} {'Relevancy %':<15}")
print("-" * 60)
for scenario, eval_result in comparison_data:
    print(f"{scenario:<20} {eval_result['score']}/5{'':<5} "
          f"{eval_result['relevant_contexts']}/{eval_result['total_contexts']}{'':<10} "
          f"{eval_result['relevancy_percentage']}%")

print("\n" + "=" * 80)
print("RAG Evaluation Complete!")
print("=" * 80)
print("\nKey Insight:")
print("- Higher scores indicate better retrieval quality")
print("- RAG systems should aim for >80% relevancy")
print("- Mixed contexts can still be useful but add noise")
```
