```python
from openai import OpenAI

client = OpenAI(api_key=api_key)
```

```python
msgs=[
        {"role": "user", "content": "Tell me a very short, simple story."},

    ]

response = client.chat.completions.create(
    model="gpt-5-nano",
    messages=msgs
)

print(response.choices[0].message.content)
```

```python
msgs = [
    {"role": "user", "content": "Tell me a very short, simple story."},
]

response = client.chat.completions.create(
    model="gpt-4o",
    messages=msgs,
    max_completion_tokens=100,     # maximum tokens in the generated response
    temperature=0.0,    # randomness: 0 = deterministic, higher = more creative
    top_p=0.9,          # nucleus sampling
)

print(response.choices[0].message.content)
```
