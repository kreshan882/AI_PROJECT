```python
msgs = [
    {"role": "user", "content": "Tell me a thitukural in tamil, simple anser only."},
]

response = client.chat.completions.create(
    model="gpt-4o",
    messages=msgs,
    max_completion_tokens=100,     # maximum tokens in the generated response
    temperature=0.0,    #  0 = original data, 1 = more creative
    top_p=0.9,          # nucleus sampling
)

print(response.choices[0].message.content)
```
